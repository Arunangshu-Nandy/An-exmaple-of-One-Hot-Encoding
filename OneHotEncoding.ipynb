```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'A':('Ram',np.nan,'Rahim','Joseph'),'B':(2,34,np.nan,56),'C':('History','Geography','Mathematics',np.nan),
                'D':(np.nan,np.nan,'Shyam','Ram')})
df
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

column_trans = ColumnTransformer(
[('A', SimpleImputer(strategy='most_frequent'), [0]),
 ('B', SimpleImputer(strategy='mean', fill_value=29), [1]),
('C',SimpleImputer(strategy='most_frequent'),[2]),
('D',SimpleImputer(strategy='most_frequent'),[3])],
remainder='passthrough')

X=pd.DataFrame(column_trans.fit_transform(df),columns=list('ABCD'))
X
```

```python
X.info()
```

```python
object_cols=[cols for cols in df.columns if df[cols].dtype=='object']
object_cols
```

```python
from sklearn.preprocessing import OneHotEncoder
OH=OneHotEncoder(handle_unknown='ignore',sparse=False)

Y=pd.DataFrame(OH.fit_transform(X[object_cols]))
Y.index=X.index
Y=pd.concat([Y,X.B],axis=1)
Y
```
